This code creates a heatmap of the distribution of damage grades across different districts in a dataset using the Pandas and Seaborn libraries for data analysis and visualization

```python
# Calculate counts
df_temp = df_stru.groupby(['district_id','damage_grade']).size().reset_index(name='count')

# Set Index
df_temp = df_temp.set_index(['district_id', 'damage_grade'])

# Calculate Proportion of grade
df_temp = df_temp.groupby(level=0).apply(lambda x: 100 * x / float(x.sum())).reset_index()

# Pivot table
df_temp = pd.pivot_table(df_temp, values='count', index=['district_id'], columns='damage_grade')

# Plot char
plt.figure(figsize=(12,5))
ax = sns.heatmap(data = df_temp, annot = True )
plt.xticks(rotation = 50)
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.title("Distribution of Damage Grade by District")
plt.xlabel("Damage Grade")
plt.ylabel("District")
plt.show()

# Clean up
del df_temp, ax
```

#### This code creates training and testing datasets for a machine learning model.

#### train_test_split is a function from scikit-learn library, which is used to split the dataset into two subsets - training and testing - for supervised learning. It takes several parameters:

#### df_stru.drop('damage_grade', axis = 1) is the feature matrix of the dataset, from which the target variable (damage_grade) is dropped.
#### df_stru['damage_grade'] is the target vector of the dataset.
#### test_size specifies the proportion of the dataset to include in the test split.
#### random_state is an integer that is used to ensure reproducibility of the results.
#### stratify parameter is set to df_stru['damage_grade'] to ensure that the target variable's class proportions are maintained in both the training and testing sets.
#### shuffle parameter is set to True to randomly shuffle the data before splitting.
#### The function returns four datasets:

#### x_train: a subset of features used for training the model.
#### x_test: a subset of features used for testing the model.
#### y_train: a subset of target variable used for training the model.
#### y_test: a subset of target variable used for testing the model.

```python
# Create training and testing data
x_train, x_test, y_train, y_test = train_test_split(df_stru.drop('damage_grade', axis = 1), 
                                                    df_stru['damage_grade'],
                                                    test_size = 0.2, 
                                                    random_state = 1989, 
                                                    stratify = df_stru['damage_grade'],
                                                    shuffle=True)
```

#### This code performs target variable processing.

#### The first step is creating a LabelEncoder object named preprocessor_tar. This object is used to transform the target variable, which is stored in the y_train and y_test variables.

#### The fit_transform() method of the preprocessor_tar object is applied to both y_train and y_test to convert the categorical values in the target variable to numerical values.

#### Then, the code creates a pandas DataFrame from y_train and uses the value_counts() method to show the counts of each target value in the training set, normalized to the total number of instances.

#### Finally, the preprocessor_tar object is deleted to clean up memory.

```python
# ------------ Target Processing ------------

# Target Variable Transformer
preprocessor_tar = LabelEncoder()
y_train = preprocessor_tar.fit_transform(y_train)
y_test = preprocessor_tar.fit_transform(y_test)

# Visualise Counts
pd.DataFrame(y_train).value_counts(normalize = True)

# Clean objects
del preprocessor_tar
```

This code performs preprocessing steps on the predictor variables in a machine learning model. The code performs the following operations:

1. Prints the shape of x_train dataframe before processing
2. Identifies the columns that are categorical, numerical, and that need to be converted to an object data type or have outlier values
3. Creates a ColumnTransformer object that will apply different preprocessing steps to different types of data, including converting data types, outlier treatment, scaling, one-hot encoding, and removing zero-variance features
4. Applies the preprocessing steps to both the x_train and x_test dataframes
5. Prints the shape of x_train dataframe after processing
6. Cleans up the intermediate objects created during the preprocessing steps

```python
# Print x train shape before processing
print("Before preprocessing there were {} rows and {} columns".format(*x_train.shape))

# ------------ Predictor Processing ------------
# Identify columns
fts_cvt_obj = ['district_id', 'vdcmun_id', 'ward_id']
fts_outlier = ['age_building']
fts_cat = df_stru.drop(fts_cvt_obj, axis = 1).select_dtypes(include=['object']).drop('damage_grade', axis = 1).columns
fts_num = df_stru.select_dtypes(np.number).columns

# Convert to object Transformer
def covert_to_object(x):
    '''Converts a column to object'''
    return pd.DataFrame(x).astype(object)
trans_to_object = Pipeline(steps = [('convert_to_object', FunctionTransformer(covert_to_object))])

# Outlier Restriction
trans_outlier = Pipeline(steps = [('Outlier_scaler', RobustScaler(quantile_range = (0,0.9)))])

# Categorical Transformer
trans_cat = Pipeline(steps = [('onehot', OneHotEncoder(handle_unknown='ignore'))])

# Numerical Transformer
trans_num = Pipeline(steps = [('scaler', StandardScaler()),
                              ('MinMax', MinMaxScaler())])

# Zero or Near Zero variance
trans_nzv = Pipeline(steps = [('nzv', VarianceThreshold(threshold = 0.01))])

# Create a single Preprocessing step for predictors
preprocessor_preds = ColumnTransformer(
    transformers=[
        ('convert_to_object', trans_to_object, fts_cvt_obj), # Convert data types
        ('Outlier', trans_outlier, fts_outlier), # Outlier treatment 
        ('num', trans_num, fts_num), # Centre and scale
        ('cat', trans_cat, fts_cat), # One Hot encode
        ('nzv', trans_nzv,[]) # One Hot encode
    ])
      
# Apply the transformations to both train and test
x_train = preprocessor_preds.fit_transform(x_train)
x_test = preprocessor_preds.fit_transform(x_test)   

# Print x train shape before processing
print("After preprocessing there are {} rows and {} columns".format(*x_train.shape))

# Clean objects
del fts_cvt_obj, fts_outlier, fts_cat, fts_num, covert_to_object, trans_to_object, trans_outlier, trans_cat, trans_num, trans_nzv
```

This code trains a random forest classifier on the training data and evaluates its performance on the test data.

1. RandomForestClassifier() creates an instance of the random forest classifier.
2. rf.fit(x_train, y_train) trains the random forest model on the training data.
3. y_pred = rf.predict(x_test) predicts the target variable on the test data.
4. confusion_matrix(y_test, y_pred) calculates the confusion matrix based on the predicted values and actual values.
5. round(100*rf.score(x_test, y_test),2) calculates the accuracy of the model on the test data.
6. round(100*rf.score(x_train, y_train),2) calculates the accuracy of the model on the training data.
7. np.sqrt(mean_squared_error(y_test,y_pred)) calculates the mean squared error (MSE) between predicted and actual target variable values.
8. sns.heatmap(con_mat, annot = True) creates a heatmap of the confusion matrix.
9. pd.DataFrame(y_test) creates a DataFrame of the actual target variable values on the test data.
10. pd.DataFrame(y_pred) creates a DataFrame of the predicted target variable values on the test data.
11. The resulting output includes the accuracy, MSE, and confusion matrix of the model, as well as the actual and predicted target variable values in separate Excel files.

```python
# Build the random forest on the full training data
rf = RandomForestClassifier() # instance
rf.fit(x_train, y_train) # fit model
y_pred = rf.predict(x_test) # predict on test

# Calculate confusion matrix
con_mat = confusion_matrix(y_test, y_pred)
con_mat = con_mat / con_mat.astype(np.float).sum(axis=1)


print('Test Accuracy of random forest Model: ',round(100*rf.score(x_test, y_test),2),'%')
print('')
print('Train Accuracy of random forest Model: ',round(100*rf.score(x_train, y_train),2),'%')
print('')

# Evaluate model

print('Mean Squared Error (MSE): ',round(np.sqrt(mean_squared_error(y_test,y_pred)),4))


# Plot Model
plt.figure(figsize = (12,6))
ax = sns.heatmap(con_mat, annot = True)
plt.title("Confusion Matrix")
plt.show()

y_test_data = pd.DataFrame(y_test)
writer = pd.ExcelWriter("y-test-RF.xlsx", engine='xlsxwriter')
y_test_data.to_excel(writer,sheet_name = "sheet1_actual_data", index=False)
writer.save()

y_pred_data = pd.DataFrame(y_pred)
writer = pd.ExcelWriter("y_pred_RF.xlsx", engine='xlsxwriter')
y_pred_data.to_excel(writer,sheet_name = "sheet1_actual_data", index=False)
writer.save()
```

This code imports the DecisionTreeRegressor model from sklearn.tree and train_test_split from sklearn.model_selection. It then creates an instance of the DecisionTreeRegressor model and fits it to the training data (x_train and y_train). After fitting the model, it calculates and prints the accuracy score on both the test and training sets, and also calculates and prints the mean squared error (MSE) between the predicted y values (y_pred) and the actual y values (y_test).

Finally, the predicted y values are stored in a Pandas DataFrame and saved as an Excel file named "y_pred_DT.xlsx".

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
reg=DecisionTreeRegressor()
reg.fit(x_train, y_train)
print('Test Accuracy of DecisionTree Regression: ',round(100*reg.score(x_test, y_test),2),'%')
print('')
print('Train Accuracy of DecisionTree Regression:',round(100*reg.score(x_train, y_train),2),'%')
print('')
y_pred=reg.predict(x_test)
print('Mean Squared Error (MSE): ',round(np.sqrt(mean_squared_error(y_test,y_pred)),4))


y_pred_data = pd.DataFrame(y_pred)
writer = pd.ExcelWriter("y_pred_DT.xlsx", engine='xlsxwriter')
y_pred_data.to_excel(writer,sheet_name = "sheet1_actual_data", index=False)
writer.save()
```
